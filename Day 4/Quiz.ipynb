(Optional) หานามสกุลที่ซ้ำกัน และแต่ละนามสกุลที่ซ้ำกันซ้ำกันกี่คน

```python
split_data = df["Name"].str.split(", ")
data = split_data.to_list()
names = ["Surname", "Firstname"]
new_df = pd.DataFrame(data, columns=names)
new_df
```

```python
df2 = new_df['Surname']
df3 = pd.DataFrame(df2.value_counts())
df3.rename(columns = {'Surname':'Counts'}, inplace=True)
df3[df3['Counts'] != 1]
```

(Optional) แสดงผลนามสกุลที่ไม่มีการซ้ำกันเลย และนับว่ามีกี่นามสกุล

```python
ans = df3[df3['Counts'] == 1]
print('มี {} นามสกุลที่ไม่ซ้ำกัน'.format(len(ans)))
ans
```
