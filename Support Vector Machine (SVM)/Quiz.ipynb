12. สร้าง train/test split ของข้อมูล สามารถลองทดสอบ 70:30, 80:20, 90:10
ratio ได้ตามใจ

```python
#Data cleaning
df.drop('id', axis=1, inplace = True)
diagnosis_result = pd.get_dummies(df['diagnosis_result'], drop_first=True)
df = pd.concat([df,diagnosis_result], axis=1)
df.drop('diagnosis_result', axis=1, inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
#Normalized data
min_max_scaler = MinMaxScaler()
df_minmax = min_max_scaler.fit_transform(df.drop('M', axis = 1))
n_df = pd.DataFrame(df_minmax, columns=df.columns[:-1])
n_df
```

```python
#Default
X_train, X_test, y_train, y_test = train_test_split(df.drop('M', axis = 1), df['M'], test_size=0.2, random_state=100)
```

```python
#Normalization
X_train2, X_test2, y_train2, y_test2 = train_test_split(n_df, df['M'], test_size=0.2, random_state=100)
```

15. เลือกช้อยที่ดีที่สุดจากข้อ 14 (หรือจะทำทุกอันแล้วนำมาเปรียบเทียบก็ได้)

16. วัดผลโมเดล โดยใช้ confusion matrix และ ประเมินผลด้วยคะแนน
Accuracy, F1 score, Recall, Precision


```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
#Default
confusion_matrix(y_test, predicted)
```

```python
#Normalization
confusion_matrix(y_test2, predicted2)
```

```python
#Default
print('Accuracy = ', accuracy_score(y_test, predicted))
print('F1 Score = ', f1_score(y_test, predicted))
print('Precision = ', precision_score(y_test, predicted))
print('Recall = ', recall_score(y_test, predicted))
```

```python
#Normalization
print('Accuracy = ', accuracy_score(y_test2, predicted2))
print('F1 Score = ', f1_score(y_test2, predicted2))
print('Precision = ', precision_score(y_test2, predicted2))
print('Recall = ', recall_score(y_test2, predicted2))
```

17. หาค่า parameter combination ที่ดีที่สุด สำหรับ Dataset นี้ โดยใช้
GridSearch (Hyperparameter Tuning)

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_combination = {'C':[0.01,0.1,1,10,100,1000], 'gamma':[0.00001,0.0001,0.001,0.01,0.1,1,10,100]}
```

```python
grid_search = GridSearchCV(SVC(), param_combination, verbose=1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_predicted = grid_search.predict(X_test)
```

```python
#Hyperparameter Tuning
print('Accuracy = ', accuracy_score(y_test, grid_predicted))
print('F1 Score = ', f1_score(y_test, grid_predicted))
print('Precision = ', precision_score(y_test, grid_predicted))
print('Recall = ', recall_score(y_test, grid_predicted))
```
