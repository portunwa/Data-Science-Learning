## 9

```python
X = df[['STG','PEG']]
X
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
fig = plt.figure(figsize =(12,8))
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('dendrogram of PEG VS STG')
```

## 12

```python
hc = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
```

```python
predicted = hc.fit_predict(X)
```

```python
fig = plt.figure(figsize=(12,8))

x1 = X[predicted == 0].iloc[:,[0]]
y1 = X[predicted == 0].iloc[:,[1]]
x2 = X[predicted == 1].iloc[:,[0]]
y2 = X[predicted == 1].iloc[:,[1]]
x3 = X[predicted == 2].iloc[:,[0]]
y3 = X[predicted == 2].iloc[:,[1]]

fig = plt.figure(figsize=(12,8))
plt.scatter(x1,y1,s=20,c='red',label = 'cluster1')
plt.scatter(x2,y2,s=20,c='blue',label = 'cluster2')
plt.scatter(x3,y3,s=20,c='black',label = 'cluster3')

plt.title('Number of cluster = 3')
plt.xlabel('STG')
plt.ylabel('PEG')
plt.legend()
```

## 13

```python
fig = plt.figure(figsize =(12,8))
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('dendrogram of PEG VS LPR')
```

## 14

```python
hc = AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
```

```python
predicted2 = hc.fit_predict(X)
```

## 15

```python
fig = plt.figure(figsize=(12,8))

x1 = X[predicted2 == 0].iloc[:,[0]]
y1 = X[predicted2 == 0].iloc[:,[1]]
x2 = X[predicted2 == 1].iloc[:,[0]]
y2 = X[predicted2 == 1].iloc[:,[1]]
x3 = X[predicted2 == 2].iloc[:,[0]]
y3 = X[predicted2 == 2].iloc[:,[1]]
x4 = X[predicted2 == 3].iloc[:,[0]]
y4 = X[predicted2 == 3].iloc[:,[1]]

fig = plt.figure(figsize=(12,8))
plt.scatter(x1,y1,s=20,c='red',label = 'cluster1')
plt.scatter(x2,y2,s=20,c='blue',label = 'cluster2')
plt.scatter(x3,y3,s=20,c='black',label = 'cluster3')
plt.scatter(x3,y3,s=20,c='yellow',label = 'cluster4')


plt.title('Number of cluster = 4')
plt.xlabel('LPR')
plt.ylabel('PEG')
plt.legend()
```
