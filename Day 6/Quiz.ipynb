```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

สร้าง DataFrame ที่เก็บยอดขายของบริษัท A,B,C,D ของปี 2014,2015,2016,2017 โดย
ชื่อบริษัทเป็น index และปีเป็น column. สร้างคำสั่ง numpy สุ่มเลขแบบ float ให้เป็นรายได้
ของบริษัทที่อยู่ในช่วง 50 – 500 (หน่วยล้านบาท)

```python
years = ['2014','2015','2016','2017',]
indexx = ['A Company','B Company','C Company','D Company']
valuess = np.random.ranf((4,4)) * np.random.randint(50,500,(4,4))
dfs = pd.DataFrame(valuess,index=indexx,columns=years)
dfs
```

ใช้ข้อมูลจาก DataFrame ด้านบนสร้าง Bar plot ให้แกน x เป็น ชื่อบริษัท ยอดขายเฉลี่ยเป็น
แกน Y โดยสร้างกราฟแบบแนวตั้งและแนวนอน

```python
aveg = [dfs.iloc[0].mean(), dfs.iloc[1].mean(), dfs.iloc[2].mean(), dfs.iloc[3].mean()]
sns.barplot(x=dfs.index,y=aveg,data=dfs)
```

```python
sns.barplot(x=aveg,y=dfs.index,data=dfs)
```

ใช้ข้อมูลจาก DataFrame ด้านบนสร้าง Scatter plot ให้แกน x เป็น ชื่อบริษัท ยอดขายปี
2017 เป็นแกน Y โดยสร้างกราฟแบบแนวตั้งและแนวนอน

```python
sns.scatterplot(x=dfs.index,y='2017', data=dfs)
```

```python
sns.scatterplot(x='2017',y=dfs.index, data=dfs)
```

สร้าง HeatMap โดย crosstab กำหนดให้ index เป็น Pclass และ column เป็นเพศ

```python
plt.figure(figsize=(5,5))
sns.heatmap(pd.crosstab(df['Pclass'], df['Sex']), cmap='coolwarm')
```

สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ
values เป็น ราคา

```python
sns.heatmap(df.pivot_table(index='Pclass',columns='Sex',values='Fare') , cmap='coolwarm')
```

สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ
values เป็น ราคา และกำหนดเส้นสีขาวขนาด 2 เป็นเส้นแบ่ง

```python
sns.heatmap(df.pivot_table(index='Pclass',columns='Sex',values='Fare') , cmap='coolwarm',linecolor='white',linewidth=2)
```
