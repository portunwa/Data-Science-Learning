12. สร้าง train/test split ของข้อมูล สามารถลองทดสอบ 70:30, 80:20, 90:10
ratio ได้ตามใจ

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
min_max_scaler = MinMaxScaler()
df_minmax = min_max_scaler.fit_transform(df.drop('Class', axis = 1))
df_minmax
```

```python
new_df = pd.DataFrame(df_minmax, columns=df.columns[1:])
new_df
```

15. เลือกช้อยที่ดีที่สุดจากข้อ 14 (หรือจะทำทุกอันแล้วนำมาเปรียบเทียบก็ได้)

16. วัดผลโมเดล โดยใช้ confusion matrix และ ประเมินผลด้วยคะแนน
Accuracy,

F1 score, Recall, Precision


```python
X_train, X_test, y_train, y_test = train_test_split(new_df, df['Class'], test_size = 0.2, random_state=100)
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,y_train)
```

```python
predicted = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
confusion_matrix(y_test, predicted)
```

```python
print('Accuracy = ', accuracy_score(y_test, predicted))
print('F1 Score = ', f1_score(y_test, predicted, average = 'micro'))
print('Precision = ', precision_score(y_test, predicted, average = 'micro'))
print('Recall = ', recall_score(y_test, predicted, average = 'micro'))
```

17. หาค่า K ที่ดีที่สุด สำหรับ Dataset นี้

```python
accuracy_lst = []
for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    predicted_i = knn.predict(X_test)
    accuracy_lst.append(accuracy_score(y_test, predicted_i))
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,50), accuracy_lst, color='black', linestyle='dashed', marker='o', markerfacecolor='red', markersize=7)
plt.xlabel('Number of K')
plt.ylabel('Accuracy')
```

```python
print(' K = 50 ดีที่สุดใน Dataset นี้')
```

18. เลือกเฉพาะ features ที่สนใจมาเทรนโมเดล และวัดผลเปรียบเทียบกับแบบ
all-features

```python
X = new_df[['Alcohol','Total phenols','Flavanoids']]
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, df['Class'], test_size = 0.2, random_state=100)
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train2,y_train2)
```

```python
predicted2 = knn.predict(X_test2)
```

```python
confusion_matrix(y_test2, predicted2)
```

```python
print('Accuracy = ', accuracy_score(y_test2, predicted2))
print('F1 Score = ', f1_score(y_test2, predicted2, average = 'micro'))
print('Precision = ', precision_score(y_test2, predicted2, average = 'micro'))
print('Recall = ', recall_score(y_test2, predicted2, average = 'micro'))
```

```python
print('เลือกบาง Features ได้ค่า Accuracy มากกว่า All-features')
```
