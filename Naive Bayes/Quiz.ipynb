15. เลือกช้อยที่ดีที่สุดจากข้อ 14 (หรือจะทำทุกอันแล้วนำมาเปรียบเทียบก็ได้)

```python
from sklearn.naive_bayes import GaussianNB
```

```python
#Default All-features
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
predicted = nb.predict(X_test)
```

```python
confusion_matrix(y_test, predicted)
```

```python
print('Accuracy = ', accuracy_score(y_test, predicted))
print('F1 Score = ', f1_score(y_test, predicted, average='micro'))
print('Precision = ', precision_score(y_test, predicted, average='micro'))
print('Recall = ', recall_score(y_test, predicted, average='micro'))
```

```python
#Normalization before splitting
mms = MinMaxScaler()
df_minmax = mms.fit_transform(df.drop(['Type'], axis=1))
norm_df = pd.DataFrame(df_minmax, columns=df.columns[:-1])
```
