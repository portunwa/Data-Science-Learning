```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน

```python
x1 = np.arange(0,21)
y1 = x1**2
plt.plot(x1,y1)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Y = X^2 function')
plt.show()
```

สร้างกราฟฟังก์ชั่น Sin (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสี
เส้นเป็นสีชมพู

```python
x2 = np.arange(0,21)
y2 = np.sin(x2)
plt.plot(x2,y2,c='#ff66ff')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Y = sin(x) function')
plt.show()
```

สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสี
เส้นเป็นสีเขียวอ่อนโดยใช้ค่ารหัสสีฐานสิบหก

```python
x21 = np.arange(0,21)
y21 = x21**2
plt.plot(y21,x21,c='#99ff99')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('X = Y^2 function')
plt.show()
```

สร้างกราฟฟังก์ชั่น Cos (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสี
เส้นเป็นสีแดง ใส่ Line Style เป็นเส้น - -

```python
x3 = np.arange(0,21)
y3 = np.cos(x3)
plt.plot(x3,y3,c='r',ls='--')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Y = cos(x) function')
plt.show()
```

สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ Y = X^3 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อ
กราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker
และเส้นสองใช้สามเหลี่ยมเป็น marker

```python
x4 = np.arange(0,21)
y41 = x4**2
y42 = x4**3
plt.plot(x4,y41,c='r', marker='o')
plt.plot(x4,y42,c='b', marker='^')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graphs function')
plt.show()
```

สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ X = Y^2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อ
กราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker
และเส้นสองใช้สามเหลี่ยมเป็น marker และใส่ legend และเพิ่มเส้นตารางข้างหลัง

```python
x5 = np.arange(0,21)
y51 = x4**2
y52 = x4**2
plt.plot(x5,y51,c='r', marker='o', label='Y = X^2')
plt.plot(y52,x5,c='b', marker='^', label='X = Y^2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graphs function')
plt.legend()
plt.grid()
plt.show()
```

สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 + 4*Y พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน
เปลี่ยนสีเส้นเป็นสีเหลืองอ่อนโดยใช้ค่ารหัสสีฐานสิบหก จำกัดแกน X และ Y ให้เหมาะสม

```python
x6 = np.arange(0,21)
y6 = x6**2 + 4*x6
plt.plot(y6,x6,c='#ffff00')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('X = Y^2 + 4*Y function')
plt.xlim([0,100])
plt.ylim([0,10])
plt.show()
```

สร้างกราฟใหญ่ด้วยสมการ X = Y^3 และกราฟเล็กด้วย Y = X^2 + 2 พล็อตลงในรูปเดียวกัน
พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน และใส่ legend

```python
x7 = np.arange(0,21)
y71 = x7**3
y72 = x7**2 + 2
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.95,0.85])
ax1.plot(y71,x7, 'r',label='X = Y^3')
plt.title('Graph functions')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
ax2 = fig.add_axes([0.6,0.4,0.4,0.3])
ax2.plot(x7,y72, 'b',label='Y = X^2 + 2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

สร้างพล็อตแบบกระจายโดยให้อายุเป็นแกน X และ ราคาตั๋วโดยสารเป็นแกน Y

สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารเฉลี่ยแต่ละชั้น

```python
first= df[df['Pclass'] == 1]['Fare'].mean()
second = df[df['Pclass'] == 2]['Fare'].mean()
third=df[df['Pclass'] == 3]['Fare'].mean()
data5 = [first,second,third]
plt.figure(figsize=[5,5])
plt.bar(['First Class','Second Class','Third Class'], data5)
plt.xlabel('Class')
plt.ylabel('Price')
plt.title('Bar chart of avg fare vs class')
plt.show()
```

สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารต่ออายุของผู้โดยสาร

```python
data6 = df['Fare'] 
data7 = df['Age']
plt.figure(figsize=[5,5])
plt.bar(data7,data6)
plt.show()
```

สร้าง bar chart โดยใช้ข้อมูลผู้ที่นามสกุลซ้ำกัน โดยความถี่เป็นแกน Y

```python
split_data = df["Name"].str.split(", ")
data = split_data.to_list()
names = ["Surname", "Firstname"]
new_df = pd.DataFrame(data, columns=names)
df2 = new_df['Surname']
df3 = pd.DataFrame(df2.value_counts())
df3.rename(columns = {'Surname':'Counts'}, inplace=True)
data8 = df3[df3['Counts'] != 1]
plt.figure(figsize=[20,5])
plt.bar(data8.index.tolist(), data8['Counts'])
plt.xlabel('Number of same surname')
plt.xticks(rotation=90)
plt.ylabel('Number of peoples')
plt.title('Bar Chart')
plt.show()

data8
```
