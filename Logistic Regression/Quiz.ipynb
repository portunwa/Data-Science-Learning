```python
#Data cleaning
df['age'].fillna(value=df['age'].mean(), inplace = True)
df['bp'].fillna(value=df['bp'].mean(), inplace = True)
df['sg'].fillna(value=df['sg'].mean(), inplace = True)
df['al'].fillna(value=df['al'].mean(), inplace = True)
df['su'].fillna(value=df['su'].mean(), inplace = True)
df['bgr'].fillna(value=df['bgr'].mean(), inplace = True)
df['bu'].fillna(value=df['bu'].mean(), inplace = True)
df['sc'].fillna(value=df['sc'].mean(), inplace = True)
df['sod'].fillna(value=df['sod'].mean(), inplace = True)
df['pot'].fillna(value=df['pot'].mean(), inplace = True)
df['hemo'].fillna(value=df['hemo'].mean(), inplace = True)

df['pcv'] = pd.to_numeric(df['pcv'])
df['wc'] = pd.to_numeric(df['wc'])
df['rc'] = pd.to_numeric(df['rc'])
df['pcv'].fillna(value=df['pcv'].mean(), inplace = True)
df['wc'].fillna(value=df['wc'].mean(), inplace = True)
df['rc'].fillna(value=df['rc'].mean(), inplace = True)

df['rbc'].fillna(value='normal', inplace = True)
df['pc'].fillna(value='normal', inplace = True)

df.dropna(inplace=True)
df.drop('id', axis=1, inplace = True)
```

```python
df.info()
```

```python
rbc = pd.get_dummies(df['rbc'], drop_first=True)
pc = pd.get_dummies(df['pc'], drop_first=True)
pcc = pd.get_dummies(df['pcc'], drop_first=True)
ba = pd.get_dummies(df['ba'], drop_first=True)
classification = pd.get_dummies(df['classification'], drop_first=True)
htn = pd.get_dummies(df['htn'], drop_first=True)
dm = pd.get_dummies(df['dm'], drop_first=True)
cad = pd.get_dummies(df['cad'], drop_first=True)
appet = pd.get_dummies(df['appet'], drop_first=True)
pe = pd.get_dummies(df['pe'], drop_first=True)
ane = pd.get_dummies(df['ane'], drop_first=True)
```

```python
df = pd.concat([df,rbc,pc,pcc,ba,classification,htn,dm,cad,appet,pe,ane], axis=1)
df.drop(['rbc','pc','pcc','ba','classification','htn','dm','cad','appet','pe','ane'], axis=1, inplace=True)
```

สร้าง train/test split ของข้อมูล สามารถลองทดสอบ 70:30, 80:20, 90:10
ratio ได้ตามใจ

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
X= df.drop('notckd', axis=1)
y = df['notckd']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)
```

ทำ Data Visualization อื่นๆ (แล้วแต่เลือก)

```python
sns.heatmap(df.isnull(), cbar = False)
```

ทำ Data Cleaning โดยการลบ หรือ fill average ขึ้นอยู่กับความเหมาะสม

```python
sns.heatmap(df.isnull(), cbar = False)
```

เลือก features ที่สนใจนำมาเทรน และ เลือก features ทั้งหมด

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train, y_train)
```

วัดผลโมเดล โดยใช้ confusion matrix และ ประเมินผลด้วยคะแนน
Accuracy, F1 score, Recall, Precision แล้วดูว่าแบบ features ที่เราเลือกมา กับ แบบเลือก
ทุก features แบบใดให้ผลลัพธ์ที่ดีกว่า

```python
predicted = logistic_regression.predict(X_test)
```

```python
print(confusion_matrix(y_test,predicted))
```

```python
print('Accuracy = ', accuracy_score(y_test,predicted))
print('F1-Score = ', f1_score(y_test,predicted))
print('Precision = ', precision_score(y_test,predicted))
print('Recall = ', recall_score(y_test,predicted))
```

ทำ Standardize ข้อมูล features ทั้งหมดก่อนเทรนโมเดล

```python
from sklearn.preprocessing import StandardScaler
sc_Xtrain = StandardScaler()
sc_Xtest = StandardScaler()

X_train2 = sc_Xtrain.fit_transform(X_train)
X_test2 = sc_Xtest.fit_transform(X_test)
```

```python
logistic_regression2 = LogisticRegression()
logistic_regression2.fit(X_train2, y_train)
```

```python
predicted2 = logistic_regression2.predict(X_test2)
```

วัดผลโมเดล โดยใช้ confusion matrix และ ประเมินผลด้วยคะแนน
Accuracy, F1 score, Recall, Precision แล้วดูว่าแบบไม่ standardize กับ แบบ
standardize แบบใดให้ผลลัพธ์ดีกว่า

```python
print(confusion_matrix(y_test,predicted2))
```

```python
print('Accuracy = ', accuracy_score(y_test,predicted2))
print('F1-Score = ', f1_score(y_test,predicted2))
print('Precision = ', precision_score(y_test,predicted2))
print('Recall = ', recall_score(y_test,predicted2))
```

เลือก features ที่สนใจ และทำ standardization, เทรนโมเดล, วัดผล และ

```python
X10 = df['age']
y10 = df['notckd']

X_train10, X_test10, y_train10, y_test10 = train_test_split(X10,y10, test_size=0.2, random_state=100)

X_train10 = np.array(X_train10).reshape(-1,1)
X_test10 = np.array(X_test10).reshape(-1,1)

sc_Xtrain10 = StandardScaler()
sc_Xtest10 = StandardScaler()

X_train10 = sc_Xtrain10.fit_transform(X_train10)
X_test10 = sc_Xtest10.fit_transform(X_test10)
```

```python
logistic_regression3 = LogisticRegression()
logistic_regression3.fit(X_train10, y_train10)
```

```python
predicted10 = logistic_regression3.predict(X_test10)
```

```python
print(confusion_matrix(y_test10,predicted10))
```

```python
print('Accuracy = ', accuracy_score(y_test10,predicted10))
print('F1-Score = ', f1_score(y_test10,predicted10))
print('Precision = ', precision_score(y_test10,predicted10))
print('Recall = ', recall_score(y_test10,predicted10))
```
