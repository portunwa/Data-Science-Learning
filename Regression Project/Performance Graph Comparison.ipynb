```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = ['Linear Regression', 'Support Vector Regression', 'Decision Tree', 'Random Forest']
y = [1069,1124,1620,1134]
```

```python
fig = plt.figure(figsize=(12,8))
sns.barplot(x=X, y=y)
plt.title('RMSE of four models (Less is better)')
plt.xlabel('Models Title')
plt.ylabel('RMSE Score')
```
